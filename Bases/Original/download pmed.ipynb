```python
import os
import requests

# URL base das instâncias
BASE_URL = "http://people.brunel.ac.uk/~mastjjb/jeb/orlib/files/"

# Pasta onde as instâncias serão salvas
PASTA = "instancias_pmed"

# Criar pasta se não existir
os.makedirs(PASTA, exist_ok=True)

for i in range(1, 41):
    nome = f"pmed{i}.txt"
    url = BASE_URL + nome
    caminho = os.path.join(PASTA, nome)

    print(f"Baixando {nome}...")

    resposta = requests.get(url)

    if resposta.status_code == 200:
        with open(caminho, "wb") as f:
            f.write(resposta.content)
        print(f"✓ {nome} salvo com sucesso!")
    else:
        print(f"✗ Erro ao baixar {nome}: HTTP {resposta.status_code}")

print("\nFinalizado! Todas as instâncias foram baixadas.")
```
